Parler DATA SAMPLE FORMATION Part 1

For each month: (NOV, DEC, JAN)
1. Read the 4 CSV files containing preprocessed parleys as pandas dataframes (derived from ndjson files 030, 031, 040, 041)
2. Select appropriate number of parleys in each dataframe
3. Merge the 4 dataframes to one final dataframe
4. Save final dataframe without index as CSV file ready to be used as input for the algorithm

Dimensions:
NOV 2020 -> 75.000 parleys * 4 files = 300.000 parleys
DEC 2020 -> 75.000 parleys * 4 files = 300.000 parleys
JAN 2021 -> 25.000 parleys * 4 files = 100.000 parleys

For months combined:
1. Select appropriate number of parleys (100.000) in each final month dataframe (NOV, DEC, JAN) 
2. Merge the 3 final month dataframes to one final complete dataframe (NOV + DEC + JAN)
3. Save complete final dataframe without index as CSV file ready to be used as input for the algorithm

Dimension of complete dataframe: (NOV + DEC + JAN)
100.000 parleys NOV + 100.000 parleys DEC + 100.000 parleys JAN = 300.000 parleys

```python
import pandas as pd 
```

```python
# NOV 2020

parler_df_030_nov = pd.read_csv('C:\\Users\\cosmi\\Desktop\\ANDREEA\\bachelors-thesis\\parler_df_030_dates_nov.csv')
print(parler_df_030_nov)

parler_df_031_nov = pd.read_csv('C:\\Users\\cosmi\\Desktop\\ANDREEA\\bachelors-thesis\\parler_df_031_dates_nov.csv')
print(parler_df_031_nov)

parler_df_040_nov = pd.read_csv('C:\\Users\\cosmi\\Desktop\\ANDREEA\\bachelors-thesis\\parler_df_040_dates_nov.csv')
print(parler_df_040_nov)

parler_df_041_nov = pd.read_csv('C:\\Users\\cosmi\\Desktop\\ANDREEA\\bachelors-thesis\\parler_df_041_dates_nov.csv')
print(parler_df_041_nov)
```

```python
parler_df_030_nov = parler_df_030_nov[75000:150000]
print(parler_df_030_nov)

parler_df_031_nov = parler_df_031_nov[0:75000]
print(parler_df_031_nov)

parler_df_040_nov = parler_df_040_nov[75000:150000]
print(parler_df_040_nov)

parler_df_041_nov = parler_df_041_nov[0:75000]
print(parler_df_041_nov)
```

```python
parler_df_nov_300000 = pd.concat([parler_df_030_nov, parler_df_031_nov, parler_df_040_nov, parler_df_041_nov], ignore_index=True)
parler_df_nov_300000
```

```python
parler_df_nov_300000.to_csv('parler_df_nov_300000.csv',index=False)
```
